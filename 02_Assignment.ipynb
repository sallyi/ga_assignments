```python
# load csv file into dataframe
import pandas as pd
csv_location = 'bikeshare.csv'
bike = pd.read_csv(csv_location, index_col='instant')
bike.head()
```

```python
#attempt to convert every value to a numeric type
bike = bike.convert_objects(convert_numeric=True)
```

```python
bike = bike.replace({'yr': {0: 2011, 1:2012},
                     'workingday': {0: 'N', 1:'Y'},
                     'holiday' : {0: 'N', 1:'Y'},
                     'weekday' : {0: 'Sun', 1:'Mon', 2:'Tue',
                                  3:'Wed', 4:'Thu', 5:'Fri',
                                  6:'Sat'}})
bike.head()
```

```python
# group by day:
daily_biking = bike.groupby('dteday')
daily_stats = daily_biking.describe()
```

```python
daily_stats.head(15)
```

```python
daily_biking['weathersit','temp','atemp','hum','windspeed','casual','registered','cnt'].mean().head()
```

```python
pd.scatter_matrix(bike[['cnt', 'atemp', 'mnth', 'yr','windspeed']], figsize=(12,12))
```

```python
# group by hour:
hourly_biking = bike.groupby(['hr'])
hourly_stats = hourly_biking.describe()
hourly_stats.head(15)
```

```python
daily_biking.cnt.mean().plot()
#shows day by day useage over 2 years
```

```python
hourly_biking.cnt.mean().plot()
# Shows general bike usage throughout the day on average
# On average, the number of riders picks up around commuting hours
```

```python
# If you separate the average hourly ridership into registered vs. casual 
# riders, the data looks very different.

hourly_biking.registered.mean().plot()
hourly_biking.casual.mean().plot()

# This plot shows that while registered ridership tends to spike during
# commuting hours, the casual ridership actually does not.
# Casual ridership tends to pick up in the afternoon and steadily drop off
# into the evening.
```

```python
working_days_pivot = bike.pivot_table(values=['casual','registered'],columns='workingday', aggfunc='mean')
working_days_pivot.plot(kind='bar')
# this indicates that casual ridership goes down on working days and registered ridership goes up 
# on working days
```
