```python
# load csv file into dataframe
import pandas as pd
csv_location = 'bikeshare.csv'
bike = pd.read_csv(csv_location, index_col='instant')
bike.head()
```

```python
# clean up data
bike = bike.replace({'yr': {0:2011, 1:2012},
                     'workingday': {0: 'N', 1:'Y'},
                     'holiday' : {0: 'N', 1:'Y'},
                     'weekday' : {0: 'Sun', 1:'Mon', 2:'Tue',3:'Wed', 4:'Thu', 5:'Fri', 6:'Sat'}})
bike.head()
```

```python
# attempt to convert every value to a numeric type
bike = bike.convert_objects(convert_numeric=True)
```

```python
#get descriptive statistics on quantitative columns
quant_cols = ['weathersit','temp','atemp','hum','windspeed','casual','registered','cnt']
bike[quant_cols].describe()
```

```python
# look for potential relationships with scattermatrix of variables of interest
# We're really only interested in relationships between ridership and other variables like time and weather
pd.scatter_matrix(bike[['cnt', 'casual','registered','atemp', 'mnth', 'yr','windspeed']], figsize=(12,12))
```

```python
# group by day:
daily_biking = bike.groupby('dteday')
#look at means for numerical values
daily_biking[quant_cols].mean().head()
```

```python
# group by hour:
hourly_biking = bike.groupby(['hr'])
hourly_stats = hourly_biking[quant_cols].describe()
hourly_stats.head(15)
```

```python
#shows day by day useage over 2 years
daily_biking.cnt.mean().plot()
```

```python
hourly_biking.cnt.mean().plot()
# Shows general bike usage throughout the day on average
# On average, the number of riders picks up around commuting hours
```

```python
# If you separate the average hourly ridership into registered vs. casual 
# riders, the data looks very different.

hourly_biking.registered.mean().plot()
hourly_biking.casual.mean().plot()

# This plot shows that while registered ridership tends to spike during
# commuting hours, the casual ridership actually does not.
# Casual ridership tends to pick up in the afternoon and steadily drop off
# into the evening.
```

```python
# See how whether or not a day is a working day may impact ridership for casual and registered riders
working_days_pivot = bike.pivot_table(values=['casual','registered', 'cnt'],columns='workingday', aggfunc='mean')
working_days_pivot
```

```python
working_days_pivot.plot(kind='bar')
# this indicates that casual ridership goes down on working days and registered ridership goes up 
# on working days.
# Overall ridership goes up on working days 
```

```python
holidays_pivot = bike.pivot_table(values=['casual','registered','cnt'],columns='holiday', aggfunc='mean')
holidays_pivot.plot(kind='bar')
# This indicates that average casual ridership goes up on holidays and average registered ridership goes down 
# on holidays. Since previous plot shows that registered riders use the bikes more on 
# working days, this could indicate that registered riders are using the bikes primarily for commuting to work
# and casual riders use them when they're not working. This makes sense with the hourly data as well.
```

```python
weather_sit_pivot = bike.pivot_table(values=['casual','registered'],columns='weathersit', aggfunc='mean')
weather_sit_pivot.plot(kind='bar')
# regardless of whether or not the rider is registered or casual, ridership seems to go down with 
# increasingly bad weather
```

```python
weather_reg_pivot = bike.pivot_table(values=['registered'],columns=['weathersit'],index=['mnth'], aggfunc='mean')
weather_reg_pivot.plot(kind='bar')
weather_cas_pivot = bike.pivot_table(values=['casual'],columns=['weathersit'],index=['mnth'], aggfunc='mean')
weather_cas_pivot.plot(kind='bar')

# casual ridership and registered ridership month-by-month (on average) divided up by type of weather
# Interestingly, registered riders don't seem to mind the snow too much in January, whereas casual riders do.
```

```python
# average monthly ridership
monthly_biking = bike.groupby(['mnth'])
monthly_biking.registered.mean().plot()
monthly_biking.casual.mean().plot()

# Both registered and casual ridership picks up in the summer
# and drops off in the winter
```

```python
# month by month over 2 years
monthly_biking = bike.groupby(['yr','mnth'])
monthly_biking.registered.mean().plot()
monthly_biking.casual.mean().plot()

# Both years average ridership increased in the summer 
# But also, average registered ridership increased overall year-over-year
```

```python
windspeed_pivot = bike.pivot_table(values=['casual','registered'],
                                          index=['windspeed'], aggfunc='mean')
windspeed_pivot.plot()
# ridership seems to decrease in high windspeeds
```

```python
humid_pivot = bike.pivot_table(values=['casual','registered'],
                                          index=['hum'], aggfunc='mean')
humid_pivot.plot()
# ridership seems to decrease in high humidity
```
