````python
#11. Show the 5 most expensive salaries for each team in the year 2014.

query = """
SELECT * FROM
(SELECT teamid, 
playerid
yearid,
salary,
rank() over (PARTITION BY teamid ORDER BY salary desc) as "salary_rank"
from salaries
where yearid = '2014') as "rank"
where salary_rank < 6
"""

pd.read_sql_query(query, cnx)
````

````python
# 12. Partition the average salaries by team and year, against year. 
# Find players that were paid more than 1 standard deviation above the average salary for that team and year. 
# Show a count by playerid.

#show the players that were paid > 1 stddev more than the average
query = """
select * from salaries s
left join
(SELECT teamid, 
yearid,
AVG(salary),
STDDEV(salary)
from salaries
group by
teamid, yearid
order by teamid, yearid) as "m" 
on s.teamid = m.teamid 
and s.yearid = m.yearid
where s.salary > m.avg + m.stddev
order by s.teamid, s.yearid
"""

#show the counts for those players per team per year
query = """
select s.teamid, s.yearid, count(playerid) from salaries s
left join
(SELECT teamid, 
yearid,
AVG(salary),
STDDEV(salary)
from salaries
group by
teamid, yearid
order by teamid, yearid) as "m" 
on s.teamid = m.teamid 
and s.yearid = m.yearid
where s.salary > m.avg + m.stddev
group by s.teamid, s.yearid
order by s.teamid, s.yearid"""


pd.read_sql_query(query, cnx)
````

````python
#13. Calculate the win percentage. convert w and g into numerics (floats) to do so. (w::numeric, for example)
query = """
select teamid, 
yearid,
w as win,
g as games,
CAST(w AS float)/CAST(g AS float) as percent_win 
from teams
order by yearid, percent_win desc"""
#query = """select * from teams"""

pd.read_sql_query(query, cnx)
````

````python
# 14. rank() the total spend by team each year against their actual rank that year. 
# Is there a correlation of spend to performance?

# rank of team spend per year

query = """
select s.*,
rank() over (partition by yearid order by s.sum)
from 
(select teamid, 
yearid, 
sum(salary) as "sum"
from salaries
group by teamid, yearid
order by teamid, yearid) as "s" """

query = """
SELECT t.teamid, t.yearid, t.rank as real_rank, r.rank as spend_rank
from teams as t
left join
(select s.*,
rank() over (partition by yearid order by s.sum)
from 
(select teamid, 
yearid, 
sum(salary) as "sum"
from salaries
group by teamid, yearid
order by teamid, yearid) as "s") as "r"
on r.teamid = t.teamid
and r.yearid = t.yearid """


df = pd.read_sql_query(query, cnx)
df.corr()
````
